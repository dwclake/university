# Image Captioning

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from transformers import pipeline
from transformers.utils import logging
logging.set_verbosity(40)
```

```python
from IPython.display import Image
```

```python
# A picture of one of my cats
!gdown 18HQQbIoZIGWaui2P91dIfcxiwg3sDKUF
photo1 = "IMG_0631.jpg"
```

```python
# A picture of a squirrel
!gdown 1sPAx3TatYfaGQOW4KeiX7PP0_111CIpk
photo2 = "C8E08343-2E7C-4A7E-8AD1-70AC31747564_4_5005_c.jpeg"
```

```python
# A picture of an abandoned hydroelectric power plant in the distance
!gdown 1vnPIO1i9_U5RENnC4dCWYKn1N9okjq09
photo3 = "1D5E788A-6172-43E3-9ABA-BB9DF12CD474_1_105_c.jpeg"
```

```python
captioner = pipeline(task="image-to-text",
                     model="nlpconnect/vit-gpt2-image-captioning")
```

```python
captioner_blip = pipeline(task="image-to-text",
                          model="Salesforce/blip-image-captioning-base")
```

```python
captioner_coco = pipeline(task="image-to-text",
                          model="ydshieh/vit-gpt2-coco-en")
```

# Cat picture

```python
display(Image(filename=photo1))
```

```python
print(captioner(photo1))
```

```python
print(captioner_blip(photo1))
```

```python
print(captioner_coco(photo1))
```
